```python
df = spark.table("workspace.default.ecommerce_bronze")
display(df.limit(10))
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("workspace.default.ecommerce_bronze_delta")
df_delta = spark.table("workspace.default.ecommerce_bronze_delta")
display(df_delta.limit(10))
```

```python
%sql
-- Get unique combinations of brand and event_type from the ecommerce table
SELECT brand, event_type
FROM workspace.default.ecommerce_bronze_delta
GROUP BY brand, event_type
```

```python
%sql
-- Get total sales for each brand
SELECT brand, 
  ROUND(SUM(price), 2) AS total_sales
FROM workspace.default.ecommerce_bronze_delta
GROUP BY brand
ORDER BY total_sales DESC
```

```python
%sql
-- generate a table of daily revenue and moving average of the last 7 days

WITH daily_revenue AS (
    SELECT
        DATE(event_time) AS event_date,
        round(SUM(price)) AS total_sales
    FROM workspace.default.ecommerce_bronze_delta
    WHERE event_type = 'purchase'
    GROUP BY DATE(event_time)
)

SELECT
    event_date,
    total_sales,
    AVG(total_sales) OVER (
        ORDER BY event_date
        ROWS BETWEEN 6 PRECEDING AND CURRENT ROW
    ) AS moving_avg_7d
FROM daily_revenue
ORDER BY event_date;
```
